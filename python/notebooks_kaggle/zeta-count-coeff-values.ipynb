```python
#count ones in class_number_2 items
df_class_number_2 = df[df.class_number.eq(2)]
print(df_class_number_2.shape[0])

column_names = []
for i in range(2,1000):
    if isprime(i):
        column_names.append('a_'+str(i))

df_class_number_2_prime = df_class_number_2[column_names]

series_cnt_1=(df_class_number_2_prime.eq(1)).T.sum()

df_class_number_2_prime = df_class_number_2_prime.assign(cnt_1 = series_cnt_1)
df_class_number_2_prime['cnt_1'].value_counts()

#df_class_number_2_prime.head(100)
```

```python
#count ones in class_number_1 items
df_class_number_1 = df[df.class_number.eq(1)]
print(df_class_number_1.shape[0])

column_names = []
for i in range(2,1000):
    if isprime(i):
        column_names.append('a_'+str(i))

df_class_number_1_prime = df_class_number_1[column_names]

series_cnt_1=(df_class_number_1_prime.eq(1)).T.sum()
df_class_number_1_prime = df_class_number_1_prime.assign(cnt_1 = series_cnt_1)

df_class_number_1_prime['cnt_1'].value_counts()

#df_class_number_2_prime.head()
```
